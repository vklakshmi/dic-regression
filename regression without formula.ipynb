```python

import numpy as np
from sklearn.linear_model import LinearRegression

Area = np.array([505,744,753,814,907,1207,1447,1533,1607,1680,1907,1957,2107,3779]).reshape((-1,1))
Price = np.array([23.1,30.85,34,36.22,39,37,59.56,52.5,52,77,74,72,109,132])

model = LinearRegression().fit(Area,Price)
print('slope , m :', model.coef_)
print('Intercept ,c :', model.intercept_)
```

```python
#prediction for built-up area 1000
Area_new = np.array([1000]).reshape((-1, 1))
print("Built-up Area :",Area_new)
y_new = model.predict(Area_new)
print("Price for Built-up Area of 1000 sqft : ", y_new , "lakhs")
```

```python
import matplotlib.pyplot as plt
max_x = np.max(Area) + 100
min_x = np.min(Price) - 100

rx = np.linspace(min_x, max_x, 1000)
ry = model.intercept_ + model.coef_ * rx 
 

plt.plot(rx, ry, color='m', label='Regression Line')
plt.scatter(Area, Price, c='red', label='Scatter Plot')
 
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```
